```python
import pandas as pd
def append_csv(source_file, target_file):
    # Read the source and target CSV files into Pandas dataframes
    source_df = pd.read_csv(source_file, sep=';')
    target_df = pd.read_csv(target_file, sep=';')

    # Append the source dataframe to the target dataframe
    appended_df = target_df.append(source_df)
    return appended_df

both = append_csv('All_500.csv', 'Paid_500.csv')
print(both)
both.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def visualize_categorical_column(data_file, column_name):
    # Read the data from the CSV file into a Pandas dataframe
    df = data_file
    plt.figure(figsize=(18, 4))
    # Create a countplot to visualize the distribution of the categorical variable
    sns.countplot(data=df, x=column_name)
    
    # Set labels and title for the plot
    plt.xlabel(column_name)
    plt.ylabel('Count')
    plt.title(f'Distribution of {column_name}')
    

    # Display the plot
    plt.show()

# Example usage
data_file = 'data.csv'  # Replace with your data CSV file path
column_name = 'status'  # Replace with the name of your categorical column

visualize_categorical_column(both, 'status')
```

```python
import pandas as pd

def count_occurrences(data_file, column_name):
    # Read the data from the CSV file into a Pandas dataframe
    df = data_file

    # Count the occurrences of each category in the column
    occurrences = df[column_name].value_counts()

    # Display the count of occurrences
    print(occurrences)

count_occurrences(both, 'status')
```

```python
df_neo = pd.read_csv('Paid_500_neo.csv', sep=';')
visualize_categorical_column(df_neo, 'status')
```
